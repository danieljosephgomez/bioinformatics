## Determine the GC-Content of some DNA 

### GC-Content is the percentage of the DNA which is G or C. 

```python
dna = 'GTCCGTCCGTCACGCTTGCGATTACTGG...'
```

```python
def gc_content(dna):
    gc = 0
    for base in dna:
        if base == 'C' or base == 'G':
            gc += 1
        return float(gc)/len(dna)
```

```python
def cg_content(dna):
    cg = [base for base in dna if base in 'CG']
    return float(len(cg))/len(dna)
```

## Reverse complement
Given some DNA, find the reverse complement:

```python
def reverse_complement(dna):
    complement = {'A':'T', 'C':'G','G':'C','T':'A'}
    
    rc = ""
    for base in dna:
        rc += complement[base]
        return "".join(reversed(rc))
```

```python
def reverse_complement(dna):
    complement = {'A':'T', 'C':'G','G':'C','T':'A'}
    return "".join([complement[base] for base in dna[::-1]])
```

# Hamming distance
Count in how many places two strings differ

```python
def hamming(lhs, rhs):
    total = 0
    for i in range(len(lhs)):
        if lhs[i] != rhs[i]:
            total += 1
        return total
```

```python
def hamming(lhs, rhs):
    return len([(x,y) for x,y in zip(lhs, rhs) if x != y])
```

# Finding motifs
Sequence motif are short, recurring patterns in DNA that are presumed to have a biological function.

```python
# Given some DNA, print all the 5-mers (k-mers)

k = 5
dna = 'GTCCCGTCCGAGGGAAATTGCGCATTCTGG'

for i in range(len(dna) - k + 1):
    print dna[i:i+k]
```

```python
k = 5
dna = 'GTCCCGTCCGAGGGAAATTGCGCATTCTGG'

motifs = Counter([dna[i:i+k] for i in range(len(dna)-k+1)])
print motifs.most_common(5)    
```

## Rabbits
Write a program that prints the nth Fibonanci number:

```python
def fibonacci(n):
    parent = 1
    child = 1
    for i in range(n):
        child, parent = parent, parent + child
    return child

for i in range(10):
    print fibonacci(i)
```
